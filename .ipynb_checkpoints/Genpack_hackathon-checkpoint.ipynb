```python
# merge the data frame
df = df1.merge(df2, how='left', on=['center_id']).merge(df3, how='left', on=['meal_id'])
```

```python
# head of merged data
df.tail()
```

```python
#Check for nulls
df.isnull().sum()
```

```python
# Get unique count for each variable
df.nunique()
```

```python
# Drop the columns as explained above
df = df.drop(['week',"id", "center_id", "meal_id"], axis = 1)
```

```python
df.head()
```

```python
# Check variable data types
df.dtypes
```

```python
# Arrange columns by data type for easier manipulation
continuous_vars = ['checkout_price',  'base_price']
cat_vars = ['emailer_for_promotion', 'homepage_featured','city_code', 'region_code','center_type', 'op_area','cuisine']
df = df[['num_orders'] + continuous_vars + cat_vars]
df.head()
```

```python
# discount amount
df['discount'] = 1 - df.checkout_price/df.base_price

# discount indicator
def dis_ind(row):
    if row['checkout_price'] < row['base_price']:
        return 1
    else:
        return 0
        
df['dis_ind'] = df.apply (lambda row: dis_ind (row),axis=1)
```

```python
# Arrange columns by data type for easier manipulation
continuous_vars = ['checkout_price',  'base_price','discount']
cat_vars = ['dis_ind','emailer_for_promotion', 'homepage_featured','city_code', 'region_code','center_type', 'op_area','cuisine']
df = df[['num_orders'] + continuous_vars + cat_vars]
df.head()
```
